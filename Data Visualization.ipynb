```python
# Plotting multiple line charts
plt.figure(figsize=(15,5))
np.random.seed(1)
summer=np.random.uniform(25,45,size=30)
winter = np.random.uniform(10,30,size=30)
rainy=np.random.uniform(20,40,size=30)
plt.plot(summer)
plt.plot(winter)
plt.plot(rainy)

plt.show()
```

```python
plt.figure(figsize=(15,5))
np.random.seed(1)
summer=np.random.uniform(25,45,size=30)
winter = np.random.uniform(10,30,size=30)
rainy=np.random.uniform(20,40,size=30)
plt.plot(summer,linestyle='-.',lw=2,color='black')
plt.plot(winter,linestyle=':',lw=2,color='red')
plt.plot(rainy,linestyle='--',lw=2,color='green')
plt.title('Temperatures over the three seasons')
plt.xlabel('Day of the month')
plt.ylabel('Temp in degree centigrade')
plt.legend(['summer','winter','rainy'],loc=3)

plt.show()
```

### Pie Chart :

```python
plt.figure(figsize=(8,8))
values=[145,400,135,130,675,234]
country=['Japan','India','Vatican city','Indonesia','China','Brazil']
plt.pie(values,labels=country,colors=['blue','red','yellow','orange','pink','purple'],shadow=True)
plt.title=('Population Density')
plt.legend(country)
plt.show()
```

### Scatter Plot :

```python
plt.figure(figsize=(8,6))
weight= np.random.uniform(60,80,20)
height_2=weight*2
plt.scatter(weight,height_2,marker='*',color='purple')
plt.title('Weight vs Height')
plt.xlabel('weight',fontsize=15)
plt.ylabel('height',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
weight_m=np.random.uniform(60,80,20)
weight_f=np.random.uniform(30,70,20)
height_m=weight_m*2
height_f=weight_f*1.5

plt.scatter(weight_f,height_f,marker='*',color='black')
plt.scatter(weight_m,height_m,marker='d',color='red')
plt.title('weight vs Height')
plt.xlabel('Weight',fontsize=15)
plt.ylabel('Height',fontsize=15)
plt.legend('Male','Female')
plt.show()
```

### Bar Chart :

```python
plt.figure(figsize=(10,8))
cities=['Delhi','Mumbai','Bangalore','Hyderabad','Chennai']
price_sqyard_in_k=[80,220,150,50,30]
plt.bar(cities,price_sqyard_in_k,color='lightblue',edgecolor='black')
plt.title("Real Estate Prices",fontsize=20)
plt.xlabel('Popular cities')
plt.ylabel('Price per square yard in thousands')
plt.show()
```

### Stacked Bar Chart :

```python
plt.figure(figsize=(10,8))
cities=['Delhi','Mumbai','Bangalore','Hyderabad','Chennai']
price_sqyard_in_K_center=[80,220,150,50,30]
price_sqyard_in_K_outskirts=[40,134,145,40,20]
plt.bar(cities,price_sqyard_in_K_center,color='darkblue',edgecolor='black')
plt.bar(cities,price_sqyard_in_K_outskirts,color='lightblue',edgecolor='black')
plt.title("Real Estate Prices",fontsize=20)
plt.xlabel('Popular cities')
plt.ylabel('Price per square yard in thousands')
plt.legend(['center','outskirts'])
plt.show()
```

## Box and Whisker Plot :

```python
# salaries of different profiles in a company :
fresher = np.round(np.random.uniform(4.5,6.5,size=10),1)
lead= np.round(np.random.uniform(6,10,size=10),1)
manager=np.round(np.random.uniform(8,15,size=10),1)
director=np.round(np.random.uniform(15,30,size=10),1)
cto=np.round(np.random.uniform(35,60,size=10),1)
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([fresher,lead,manager,director,cto],labels=['fresher','lead','manager','director','cto'],patch_artist=True,vert=0)
plt.show()
```

```python
# sample box n whisker:
data=[-10,2,3,4,5,6,6,10]
plt.boxplot(data,vert=0)
plt.axvline(2)
plt.axvline(10)
plt.axvline(2.5,color='red')
plt.axvline(4.5,color='red')
plt.axvline(6,color='red')
plt.show()
```

## Histogram :
### - It is drawn for the data which is continuous and numerical.

```python
values=np.random.rand(100)*100
plt.figure(figsize=(10,5))
plt.hist(values,histtype='bar',edgecolor='black',bins=20)
plt.title('Histogram')
plt.show()
```

## Summary : 

#### - Line chart : time series and sequential data ( numerical variable )
#### - Pie chart : compare distribution of categories ( categorical variable )
#### - Bar chart : same as pie but a better version 
#### - Box and Whisker : To check for outliers and understand the data distribution ( numerical variable )
#### - Scatter : To understand the relation b/w two numerical variables i.e., checking for correlation.
#### - Histogram : To see distribution of numerical data ( shape of data and spread of data )
