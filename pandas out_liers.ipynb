```python
# outliers for loan amount column
df['LoanAmount'].head()
```

```python
df['LoanAmount'].isnull().sum()
```

```python
Q1=np.quantile(df['LoanAmount'],0.25)
Q2=np.quantile(df['LoanAmount'],0.50)
Q3=np.quantile(df['LoanAmount'],0.75)
```

```python
def outlier_detection(col,df_name):
    Q1=np.quantile(col,0.25)
    Q2=np.quantile(col,0.50)
    Q3=np.quantile(col,0.75)
    # calculating IQR
    IQR = Q3-Q1
    # lower boundary
    Lower_boundary = Q1 - (1.5*IQR)
    # upper boundary 
    Upper_boundary = Q3 + (1.5*IQR)
    result = df_name.loc[(col<Lower_boundary)|(col>Upper_boundary)]
    print(f" Total number of outliers in {col} is {result}")
    return result
```

```python
outlier_detection(df['LoanAmount'],df)
```
